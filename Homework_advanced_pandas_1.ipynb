```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
log['source_type'] = log['traffic_source']
log.loc[log.traffic_source.isin(['yandex', 'google']), 'source_type'] = 'organic'
log.loc[(log.region.str.contains('Russia')) & log.traffic_source.isin(['paid', 'email']), 'source_type'] = 'ad'
log.loc[(log.traffic_source.isin(['paid', 'email'])) & (log['region'] != 'Russia'), 'source_type'] = 'other'

log.head()
```

### Задание 3

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
```

```python
data
```

```python
def traffic_set(row):
  
    traffic = row['keyword']
    
    lemmas = m.lemmatize(traffic)
    lemmas = ' '.join(lemmas).strip(' \n')
    return lemmas
```

```python
data['lemmas'] = data.apply(traffic_set, axis=1)
data.head()
```
